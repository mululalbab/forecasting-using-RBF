```python
def normalization(dataset): # fungsi normalization dg data sebagai input 
    (baris,kolom) = dataset.shape
    mins          = np.zeros(shape = (kolom)) # inisialisasi array dataset tesis minimum
    maxs          = np.zeros(shape = (kolom)) # initiasliasi array dataset tesis maksimum
    for j in range (kolom):
        mins[j] = np.min(dataset[:,j]) # mendapatkan nilai minimum setiap kolom
        maxs[j] = np.max(dataset[:,j]) # mendapatkan nilai maksimum setiap kolom
        
    normalisasi = np.copy(dataset) #mendefinisikan hasil normalisasi 45x3
    for i in range (baris):
        for j in range (kolom):
            normalisasi[i,j] = ((dataset[i,j]-mins[j]) / (maxs[j]-mins[j])) # hasil subtitusi ke dalam rumus normalisasi (Dubey, 2015 
            np.set_printoptions(precision=11, suppress=True) # menampilkan 11 digit angka dibelakang koma               
    return (normalisasi) # display normalisasi
```

## Parameter Centers (Nilai Tengah Nilai Tengah)

```python
l =3 # jumlah clusters dan jumlah centers setiap atribut input
def ChooseICenters(ds_training, l):#fungsi ChooseICenters dg inputs ds_training 45 x 3 & l (jumlah clusters atau panjang centers)
    # mengambil centers awal dr dataset training scr acak
    centers = ds_training[np.random.choice(ds_training.shape[0], l, replace = False)]
    np.set_printoptions(precision=11, suppress=True)
    return(centers) # ukurannya l x 3 (kolom 3 karena jml atribut input 3)

def EuclideanDistance(unit, center): # fungsi Euclidean Distance dg inputs unit & center 
    baris = len(unit)
    sum   = 0
    for i in range(baris):
        sum += (unit[i] - center[i]) ** 2
    return np.sqrt(sum)

def AssigntoClusters(ds_training, centers, clustering): # fungsi Assign to Clusters dg inputs ds_training, centers & clustering
    #diberikan array initial centers, ds_training akan terkluster di clusters
    clusters           = np.copy(clustering) # membuat sama ukuran array clusters baru dg input clustering
    (baris, kolom)     = ds_training.shape
    p                  = len(centers)
    
    EuclideanDistances = np.zeros(shape = (p), dtype = np.float32)# inisialisasi array Euclidean Distances ukuran (30, :)
    for i in range(baris): # i looping terus utk setiap ds_training
        for k in range(p): # k looping terus utk setiap centers
            EuclideanDistances[k] = EuclideanDistance(ds_training[i], centers[k]) #jarak dr ds_training ke setiap centers
        criterion    = np.argmin(EuclideanDistances) #menentukan setiap ds_training berada pada clusters tertentu sesuai kriteria jarak terdekat dengan centers
        clusters[i]  = criterion #setiap ds_training berclusters
            
    if np.array_equal(clustering, clusters): # array & elements clustering = array & elements clusters
        return False # saat kondisi berhenti (tidak ada perubahan terjadi) adalah salah
    
    calc = np.zeros(shape = (p), dtype = np.int) # inisialisasi array calc ukuran (p, :)
    for i in range(baris): # i looping terus utk setiap ds_training
        number = clustering[i]
        calc[number] += 1 #ukuran array (p,:) dan #memastikan tak ada clusters yang tidak terisi (kosong) atau jumlahnya = 0 dari anggota ds_training
    
    for k in range(p):  #menghitung ada atau tidaknya isi clusters (= np.count_nonzero)
        if calc[k] == 0: #clustering jelek jika ada yg bernilai nan=0 
            return False # saat kondisi ada cluster yg tak terisi (kosong) atau = 0 adalah salah
        
    for i in range(baris):
        clustering[i] = clusters[i] #clustering = clustering baru terisi
    return True # ada perubahan clustering = clusters
    #terjadi perubahan dan tidak ada cluster yg tak terisi (kosong) atau = 0
    #memperbaharui clustering = clusters

def RecomputeCenters(ds_training, centers, clustering):
    #ada clusters baru, akan menghitung centers baru
    #clusters terdapat di function AssigntoClusters dg tak ada clusters yg kosong atau = 0
    (baris, kolom) = ds_training.shape
    p              = len(centers)
    calc           = np.zeros(shape = (p), dtype = np.int) #inisialisasi jumlah ds_training telah terkluster di setiap clusters (p,:)
    newcenters     = np.zeros(shape = centers.shape, dtype = np.float32) # ukuran array sama dg centers (p x 3)
    for i in range(baris):
        number        = clustering[i]
        calc[number] += 1
        for j in range(kolom):
            newcenters[number,j] += ds_training[i,j] #akumulasi sigma ds_training
            
    for k in range(p):
        for j in range(kolom):
            newcenters[k,j] /= calc[k] #rumus temporary centers (step 4 Gupta Book) dan calc ~= 0 #61
    #rata-rata #calc = banyaknya ds_training di dalam setiap cluster ke-k
            
    for k in range(p): # looping sebanyak panjang array centers
        for j in range(kolom):
            centers[k,j] = newcenters[k,j] #temporary centers = newcenters (terupdate)

#menjalankan algoritma ke-4 dalam Gupta utk mendapatkan  centers dari rata-rata baru            
def TemporaryCenters(ds_training, p):
    (baris, kolom) = ds_training.shape
    clustering     = np.zeros(shape = (baris), dtype = np.int) #inisialisasi clustering ukuran array (p,:)
    centers        = np.zeros(shape = (p,kolom), dtype = np.float32)
    for k in range(p): # k looping terus utk setiap centers 
        clustering[k] = k
    for k in range(p, baris):
        clustering[k] = np.random.randint(0,p) #numbers klusters sebagai atribut: [0,p-1]; ukuran: 1x45 elemen2nya [0,p-1]
    RecomputeCenters(ds_training, centers, clustering) # running fungsi RecomputeCenters 
    return(clustering, centers)

def UpdatingCenters(ds_training, p):
    (clustering, centers) = TemporaryCenters(ds_training, p) #memanggil hasil dr step 1 s.d step 4 melalui fungsi RecomputeTCenters
    
    term         = True #asumsi benar jika ada sedikitnya unit ds_training berpindah ke lain cluster (clustering jelek)
    maxiteration = 100 #batas jumlah iterasi maksimum yg diterapkan
    epoch        = 1
    while epoch < maxiteration:
        epoch += 1 
        term = AssigntoClusters(ds_training, centers, clustering) #krn benar ada perpindahan, memasukkan ds_training ke cluster baru dg fungsi AssigntoClusters
        if term == False: #sudah tidak ada perpindahan ds_training ke cluster lain lagi
            break #selesai (prosedur berhenti)
        RecomputeCenters(ds_training, centers, clustering) #updating centers hingga tak ada lagi perpindahan ds_training ke cluster lain
    return(clustering, centers, epoch)
```

## Parameter Sigma (Standar Deviasi/Spread)

```python
def sigma(centers, p):
    p              = len(centers)
    
    dmax          = 0
    for i in range(p):
        for j in range(p):
            if i == j:
                continue
            dist = EuclideanDistance(centers[i], centers[j])
            if dist > dmax:
                dmax = dist 
    sigma = dmax / np.sqrt(2*p)
    return sigma
```

## Fungsi Aktivasi Radial Basis 

```python
def gaussian(ds_training, centers, sigma):
    return np.exp(- EuclideanDistance(ds_training, centers)**2 / (2 * sigma**2))

def multiQuadratic(ds_training, centers, sigma):
    return np.power(EuclideanDistance(ds_training, centers)**2 + sigma **2, 0.5)

def inversMultiQuadratic(ds_training, centers, sigma):
    return np.power(EuclideanDistance(ds_training, centers)**2 + sigma **2, -0.5)
```

## Parameter Phi

```python
def Phi(ds_training, centers, sigma):
    (baris, kolom) = ds_training.shape
    p              = len(centers)
    
    phi            = np.zeros(shape = (baris,p), dtype=np.float32)
    for i in range (baris):
        for j in range (p):
            phi[i,j] = gaussian(ds_training[i], centers[j], sigma)              #Gaussian Radial Basis Function
            #phi[i,j] = multiQuadratic(ds_training[i], centers[j], sigma)       #multiQuadratic RBF
            #phi[i,j] = inversMultiQuadratic(ds_training[i], centers[j], sigma) #inversMultiQuadratic RBF
            np.set_printoptions(precision=11, suppress=True)      
    return(phi)
```

```python
def error(phi, da_y, w):
    (baris, kolom)     = da_y.shape
    p                  = len(w)
    e                  = np.zeros(shape=(da_y.shape), dtype=np.float32)
        
    sum3 = 0                 
    for i in range (baris):
        sum2 = 0             
        for j in range (p):
            sum2 += phi[i,j] * w[j]     
        e[i] = da_y[i] - sum2
        np.set_printoptions(precision=11, suppress=True)
        sum3 += e[i] ** 2         
    E = 0.5 * sum3               
    np.set_printoptions(precision=11, suppress=True)
    return(e, E)
```

```python
def Update_w_and_errors(phi, da_y, w, e, E):
    (baris,kolom) = da_y.shape
    p             = len(w)
    epoch         = 1
    alpha         = 0.01 #nilai learning rate
    
    maxiteration  = 5000
    while epoch < maxiteration and E > 0.1: #8
        for k in range (p):
            sum4 = 0
            for i in range (baris):
                sum4 +=  e[i] * phi[i,k]   #12
            w[k] += alpha * sum4          #13
            np.set_printoptions(precision=11, suppress=True)
            
        sum6 = 0
        for i in range (baris):
            sum5 = 0 
            for k in range (p):
                sum5 += phi[i,k] * w[k]  #20
            e[i]  = da_y[i] - sum5  #21
            np.set_printoptions(precision=11, suppress=True)
            sum6 += e[i] ** 2             #update error e #23
        E = 0.5 * sum6                    #update error E  #24
        np.set_printoptions(precision=11, suppress=True)
        epoch += 1
    return(w, e, E, epoch)
```

```python
def Approximation_y(ds_training, centers, sigma, phi, w_update):
    (baris, kolom) = ds_training.shape
    p              = len(centers)
    F              = np.array(np.zeros([baris,1]), dtype=np.float32)
        
    sum7          = 0 #sum7
    for i in range (baris):
        for j in range (p):
            phi[i,j] = gaussian(ds_training[i], centers[j], sigma)              #Gaussian Radial Basis Function
            #phi[i,j] = multiQuadratic(ds_training[i], centers[j], sigma)       #multiQuadratic RBF
            #phi[i,j] = inversMultiQuadratic(ds_training[i], centers[j], sigma) #inversMultiQuadratic RBF
            sum7   += phi[i,j]*w_update[j]             
            np.set_printoptions(precision=11, suppress=True)
        F[i,0]       = sum7                                   
        sum7        = 0   
    return (F, phi, w_update)
```

```python
def denormalization(day, approx_y): 
    (baris,kolom)  = approx_y.shape
    mins           = np.zeros(shape=(kolom))
    maxs           = np.zeros(shape=(kolom))
    
    for jj in range (kolom):
        mins[jj] = np.min(day[:,jj])
        maxs[jj] = np.max(day[:,jj])
    
    hasil = np.copy(approx_y) #mendefinisikan hasil denormalisasi 45x3
    for i in range (baris):
        for jj in range (kolom):
            hasil[i,jj] = mins[jj] + (approx_y[i,jj]) * (maxs[jj]-mins[jj])
            np.set_printoptions(precision=11, suppress=True)                 
    return (hasil)
```

```python
denormalize_ay = denormalization(day, approx_y) #menampilkan hasil dari fungsi denormalization  
print(denormalize_ay)
```

```python
def MAPE(day, denormalize_ay):
    mapeRBF = np.array(np.zeros([1,1]), dtype=np.float32)
    mapeRBF = np.mean(np.absolute(day - denormalize_ay)/day) * 100
    
    return(mapeRBF)
```
